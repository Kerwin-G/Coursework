```python
#Calculating Sequence Similarity

REV_COMP = {'G': { 'G':-1, 'C': 1, 'A':-1, 'T':-1 },
 'C': { 'G': 1, 'C':-1, 'A':-1, 'T':-1 },
 'A': { 'G':-1, 'C':-1, 'A':-1, 'T': 1 },
 'T': { 'G':-1, 'C':-1, 'A': 1, 'T':-1 }}  #Dictionary: matrix for point system

def calcSeqSimilarity(seqA, seqB, simMatrix): #Function accepts two sequences in question and matrix for assessment
 numPlaces = min(len(seqA), len(seqB)) #assess up to the length of the smallest sequence
 totalScore = 0.0
 for i in range(numPlaces): #loop for assessment up to length of shortest sequence
   residueA = seqA[i]
   residueB = seqB[i]
   totalScore += simMatrix[residueA][residueB] 
   return totalScore

print(calcSeqSimilarity('ACGTTTAAGGCATGACCA', 'TGTAAATTCGGTTCTGGT', REV_COMP)) #Return score
```

```python
#Multiple sequence alignment, returns frequency of each amino acid
def profile(alignment):
 n = len(alignment[0])
 nSeq = float(len(alignment))
 prof = []
 for i in range(n):
   counts = {}
   for seq in alignment:
     letter = seq[i]
     if letter == '-':
       continue
     counts[letter] = counts.get(letter, 0) + 1
   for letter in counts:
     counts[letter] /= nSeq
   prof.append(counts)
 return prof
alignment = ['SRPAPVVIILIILCVMAGVIGTILLISYGIRLLIK',
 'TVPAPVVIILIILCVMAGIIGTILLISYTIRRLIK',
 'HHFSEPEITLIIFGVMAGVIGTILLISYGIRRLIK',
 'HEFSELVIALIIFGVMAGVIGTILFISYGSRRLIK']
print(profile(alignment))
# First sub-dict: {'H': 0.5, 'S': 0.25, 'T': 0.25}
```
