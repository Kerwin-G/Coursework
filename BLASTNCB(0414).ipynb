```python
from Bio.Blast import  NCBIWWW
result_handle = NCBIWWW.qblast("blastn", "nt", "8332116") 
with open("my_blast.xml","w") as out_handle:
  out_handle.write(result_handle.read())
result_handle.close()
```

```python
from Bio.Blast import NCBIXML
result_handle= open("my_blast.xml")
blast_record = NCBIXML.read(result_handle)

E_VALUE_THRESH = 0.04

blast_results = {}
hits=[]
for alignment in blast_record.alignments:
  hits.append(alignment.hit_id) #number of hits in an array
  for hsp in alignment.hsps:
    if hsp.expect < E_VALUE_THRESH:
      print("****Alignment****")
      print("sequence:", alignment.title)
      print("length:", alignment.length)
      print("e_value:", hsp.expect)
      print(hsp.query[0:75] + "...")
      print(hsp.match[0:75] + "...")
      print(hsp.sbjct[0:75] + "...")
      for i in alignment.hsps:
        blast_result[i]={ #creating a dictionary of lists
            'sequence' : [alignment.title],
            'length' : [alignment.length],
            'e_value' : [hsp.expect],
            'hsp_query' : [hsp.query],
            'hsp-match' : [hsp.match],
            'hsp-subject' : [hsp.sbjct]
            }
        blast_results.update(blast_result) #push element to update dictionary

print(" ")
print(blast_record.alignments[0].hit_id) #valid!
print(blast_results)
print(hits)
print(len(hits))
print(type(blast_record))
```
